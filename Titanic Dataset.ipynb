```python
print("Number of surviving male:", sum((data["Survived"] == 1) & (data["Sex"] == "male")))
print("Number of Dead male:", sum((data["Survived"] == 0) & (data["Sex"] == "male")))
```

```python
male_survival_counts = data[data["Sex"] == "male"]["Survived"].value_counts()

labels = ["Did not survive", "Survived"]

plt.figure(figsize=(3, 3))
plt.pie(male_survival_counts, labels=labels, autopct='%1.1f%%', colors=["red", "green"], startangle=90)
plt.title("Survival Distribution of Male Passengers")
plt.show()
```

```python
print("Number of surviving Female:",sum((data["Survived"]== 1) & (data["Sex"] == "female")))
print("Number of Dead Female:",sum((data["Survived"] == 0) &(data["Sex"]== "female")))
```
